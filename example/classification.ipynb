```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from rapidpredict.supervised import rapidclassifier

data = load_breast_cancer()
X = data.data
y= data.target
# print(y)

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.25,random_state =123)

clf = rapidclassifier(verbose= 0,ignore_warnings=True, custom_metric=None)
# print("clf" ,clf)
models , predictions = clf.fit(X_train, X_test, y_train, y_test)
```

```python
from rapidpredict.supervised import plot_target
plot_target(y)
```

```python
from rapidpredict.supervised import compareModels_bargraph
compareModels_bargraph(predictions["F1 Score"] ,models.index)
```

```python
from rapidpredict.supervised import compareModels_boxplot
compareModels_boxplot(predictions["F1 Score"] ,models.index)
```
